<u>THE THREE BRANCHES OF MACHINE LEARNING</u>

Machine Learning is mainly divided into three important branches. Each branch is designed to solve different types of problems based on the nature of the data and the goal of the task.

<u>SUPERVISED LEARNING</u>

Supervised Learning is a method in which the dataset contains both inputs and their correct outputs, known as labels. The model is trained using these input–output pairs so that it can understand how the inputs are connected to the outputs. During training, the algorithm learns the mapping between them. Once the training is complete, it can predict outcomes for new, unseen data.

This type of learning is mainly used for classification and regression problems.

Examples:

Spam detection, where emails are labeled as “spam” or “not spam.”

Predicting house prices using historical data where the prices are already known.

Predicting student marks based on previous performance records.

Supervised learning works best when accurate labeled data is available.

<u>UNSUPERVISED LEARNING</u>

Unsupervised Learning is used when the dataset does not contain labeled outputs. In this case, the model does not know the correct answers in advance. Instead, it analyzes the data to discover hidden patterns, similarities, or structures on its own.

This type of learning is mainly used for clustering and association tasks.

Examples:

Customer segmentation, where customers are grouped according to purchasing behavior.

Grouping similar news articles based on content.

Identifying hidden trends in large datasets.

Unsupervised learning is helpful when we want to explore data and uncover meaningful insights without predefined labels.

<u>REINFORCEMENT LEARNING</u>

Reinforcement Learning is a learning approach where the model learns through interaction with an environment. It makes decisions, receives feedback in the form of rewards or penalties, and gradually improves its strategy to maximize rewards over time.

This method follows a trial-and-error process and is commonly used in dynamic environments.

Examples:

A game-playing AI that improves its strategy by playing repeatedly and earning rewards for winning.

A robot learning how to walk by adjusting movements based on feedback.

Self-driving systems improving decisions through environmental interaction.

Reinforcement learning is particularly useful in situations where actions directly influence future outcomes.

### <u>Why Choosing the Wrong Type is a Mistake</u>
Selecting the wrong type of Machine Learning can lead to incorrect results and poor model performance. Each branch is designed for a specific purpose, and using an unsuitable method creates confusion and inefficiency.

For example:

Supervised learning requires labeled data. If labels are not available, the model cannot learn correctly.

Unsupervised learning is used to discover patterns and group data, but it cannot provide precise predictions like supervised learning.

Reinforcement learning is suitable for environments where decisions are rewarded or penalized. Applying it to simple prediction problems would be unnecessary and ineffective.

Choosing the wrong approach can waste time, computational resources, and effort. It may also reduce accuracy and produce unreliable outcomes. Therefore, understanding the nature of the problem and the type of data available is essential before selecting a Machine Learning technique. Making the correct choice ensures better performance, efficiency, and meaningful results.
